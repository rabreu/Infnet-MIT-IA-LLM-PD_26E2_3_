```python
from transformers import AutoModel, AutoTokenizer

# Definindo o modelo (pode ser trocado dinamicamente)
model_id = "microsoft/deberta-v3-xsmall"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModel.from_pretrained(model_id)

# Processando a entrada
inputs = tokenizer("O mecanismo de atenção é poderoso", return_tensors="pt")
outputs = model(**inputs)
```

```python
print(f"Dimensões do output: {outputs.last_hidden_state.shape}")
# Resultado esperado: [Batch, Tokens, Hidden_Dim]
```

```python
qa_pipe = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
context = """
Câmara dos EUA aprova regra que impede Trump de ordenar ataque ao Irã
"""

resposta = qa_pipe(question="Existe uma guerra?",
                   context=context)
print(f"Resposta QA: {resposta['answer']}")
```
